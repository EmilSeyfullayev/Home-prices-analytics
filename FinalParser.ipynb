### Links parser

```python
product_links = []
vipped_featured = []

def parser(end_page, start_page=1):

  for page_number in range(start_page, end_page+1):

    try:

      url = f'XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX'

      html = requests.get(url).content
      soup = BeautifulSoup(html, "html.parser")

      items = list(soup.find_all("div", {'class':'items-i'}))

      print(page_number, " out of ", str(end_page))

      #time.sleep(np.random.randint(1, 4))

      for item_index in range(len(items)):
        
        product_link = 'XXXXXXXXX'.format(items[item_index].find('a')['href'])

        product_links.append(product_link)

        if "".join(items[item_index]['class']).lstrip('items-i') == "":

          vipped_featured.append('0')
        else:
          vipped_featured.append("".join(items[item_index]['class']).lstrip('items-i'))        

    
    except:
      product_links.append('0')

parser(end_page, start_page)
```

### Products parser

```python
ids = [i.lstrip("XXXXXXXXXXXXXXXXXXXXXXXXXXXXX") for i in product_links]
prices = []
texts = []
categories = []
floors = []
areas = []
rooms = []
documents = []
credits = []
address= []
district = []
latitudes = []
longitudes = []
ownerships = []
names = []
emails = []
phone_numbers_appened = []
ad_number = []
watches = []
ad_refreshed_date = []
agency_titles = []
photo_counts = []
phot_pseudo_links = []
deal_ended = []

def product_parser(product_links):

  for product_link in product_links:

    #time.sleep(np.random.randint(1, 4))    

    product_html = requests.get(product_link).content
    product_soup = BeautifulSoup(product_html, "html.parser")
  
    try:
      basa_catib = product_soup.find('p', {'class':'flash'}).text
      deal_ended.append(basa_catib)
    except:
      deal_ended.append('0')


    try:
      price = product_soup.find('span', {'class':'price-val'}).text
      prices.append(price)
    except:
      prices.append("0")

    try:
      text = product_soup.find('article').text
      texts.append(text)
    except:
      texts.append('0')

    try:
      kateqoriya = product_soup.find_all('td')[1].text
      categories.append(kateqoriya)
    except:
      categories.append('0')

    try:
      mertebe = product_soup.find_all('td')[3].text
      floors.append(mertebe)
    except:
      floors.append('0')

    try:
      sahe = product_soup.find_all('td')[5].text
      areas.append(sahe)
    except:
      areas.append("0")
    
    try:
      otaq_sayi = product_soup.find_all('td')[7].text
      rooms.append(otaq_sayi)
    except:
      rooms.append('0')

    try:
      cixaris = product_soup.find_all('td')[9].text
      documents.append(cixaris)
    except:
      documents.append('0')

    try:
      ipoteka = product_soup.find_all('td')[11].text
    except IndexError:
      ipoteka = str('xeyr')
    
    credits.append(ipoteka)

    try:
      unvan = product_soup.find('div', {'class': 'map_address'}).text
      address.append(unvan)
    except:
      address.append('0')

    rayon = []

    try:
      for i in product_soup.find_all('ul', {'class':'locations'}):
        for n in i.find_all("a"):
          rayon.append(n.text)

      final_rayon = '; '.join(rayon)
      district.append(final_rayon)
    except:
      district.append('0')

    try:
      latitude = product_soup.find('div', {'id':'item_map'})['data-lat']
      latitudes.append(latitude)

      longitude = product_soup.find('div', {'id':'item_map'})['data-lng']
      longitudes.append(longitude)
    except:
      latitudes.append('0')
      longitudes.append('0')


    try:
      ownership = product_soup.find('span', {'class':'ownership'}).text
      ownerships.append(ownership)
    except:
      ownerships.append('0')

    try:
      name_of_seller = product_soup.find('div', {'class':'name'}).text.split(ownership)[0]
      names.append(name_of_seller)
    except:
      names.append('0')
    
    try:
      email = product_soup.find('span', {'class':'email'}).text
      emails.append(email)
    except:
      emails.append(str(0))

    try:
      phone_number_url = f'{product_link}/phones'
      html_phone_number = requests.get(phone_number_url).content
      phone_numbers = BeautifulSoup(html_phone_number, 'html.parser')
      phone_number = "; ".join(json.loads(phone_numbers.text)['phones'])
      phone_numbers_appened.append(phone_number)
    except:
      phone_numbers_appened.append("0")

    try:
      elanin_nomresi = product_soup.find('div', {'class':"item_info"}).find_all("p")[0].text.split(": ")[1]
      ad_number.append(elanin_nomresi)
      baxislarin_sayi = product_soup.find('div', {'class':"item_info"}).find_all("p")[1].text.split(": ")[1]
      watches.append(baxislarin_sayi)
      yenilenme_tarixi = product_soup.find('div', {'class':"item_info"}).find_all("p")[2].text.split(": ")[1]
      ad_refreshed_date.append(yenilenme_tarixi)
    except:
      ad_number.append('0')
      watches.append('0')
      ad_refreshed_date.append('0')


    try:
      agency_title = product_soup.find('h1', 'agency--title').text
    except:
      agency_title = str('0')
   
    agency_titles.append(agency_title)
    
    try:
      photo_count = int(product_soup.find('div', class_="thumbnails")['data-count']) + 1
      photo_counts.append(photo_count)
    except:
      photo_counts.append('0')

    # to access photos we need to add 'XXXXXXXXXXXXXXXXXXXXXXXXXX'
    photo_links = []

    try:
      for i in product_soup.find_all('div', class_="thumbnail"):
        photo_links.append(i['data-mfp-src'].strip('XXXXXXXXXXXXXXXXX'))

    # ['XXXXXXXXXXXXXXXXXXXXXXXXX' + i for i in photo_links_merged.split("; ")] 

      photo_links_merged = ";".join(photo_links)#.split(";")
      phot_pseudo_links.append(photo_links_merged)
    except:
      phot_pseudo_links.append('0')


    print((product_links.index(product_link)+1), ' out of ', len(product_links))

product_parser(product_links)
```
